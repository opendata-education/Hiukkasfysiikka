## Säilyvät suureet hiukkasfysiikassa


Kaikki ovat jossain vaiheessa nähneet Einsteinin kuuluisan kaavan  $E = mc^2$, mutta monet eivät ymmärrä sen merkitystä. Einsteinin suuri löydös oli, että todellisuudessa energian ja massan välillä on yhteys ja tietyissä prosesseissa energia voi muuttua massaksi ja toisinpäin. Tämä kaava siis sanoo, että jos 1 kg massa muuttuu energiaksi, saadaan $c^2$ joulea, missä $c$ on valonnopeus.

Koska massa ei ole enää säilyvä suure, tarvitsemme uuden. Määritellään uusi suure, joka on ikään kuin kappaleen massan ja energian summa, invariantti massa. Olkoon nyt hiukkasen kokonaisenergia  $E = Mc^2 + E_{kin}$. Kineettinen energia saadaan liikemäärästä relativistisissa nopeuksissa kaavalla $E_{kin} = c^2 |p|^2$. Liike-energia lasketaan liikemäärästä, koska se on helppo mitata hiukkasille. Tästä saadaan invariantti massa kaavalla:


$$
M c^2 = \dfrac{E^2}{c^2} - |\vec{p}|^2
$$

Vastaava kaava kahden hiukkasen yhteiselle invariantille massalle on:

$$
Mc^2 = \sqrt{(E_1 + E_2)^2 - c^2 |\vec{p}_1 + \vec{p}_2|^2 }
$$
